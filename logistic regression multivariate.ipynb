```python
from numpy import *

import pandas as pd
import math
```

```python
def h_theta_avg_error(theta_0,theta_1,theta_2,data_set,m,x,x1,y):
    total_error=0
    for i in range(0,m-1):
        z=(theta_0+(theta_1*x[i])+(theta_2*x1[i]))
        hyp=(1/(1+math.e**(-1*z)))
        print(hyp)
        error=((((-1*y[i]))*(math.log(hyp)))-((1-(y[i]))*(math.log(1-hyp))))**2
        total_error+=error
    return((total_error/(m)))
```

```python
def gradient_descent(total_error,alpha,data_set,theta_0,theta_1,theta_2,iterations,x,x1,y):
    m=len(data_set)
    print("running gradient Descent....")
    i=0
    for i in range(0,iterations):
        theta_0,theta_1,theta_1,new_error=descent_step(theta_0,theta_1,theta_2,data_set,alpha,total_error,x,x1,y)
    total_error=new_error
    plot(theta_0,theta_1,theta_2,data_set,m,x,x1,y)
    return (theta_0,theta_1,theta_2,total_error,alpha)     
```

```python
def descent_step(theta_0,theta_1,theta_2,data_set,alpha,total_error,x,x1,y):
    theta_0_temp=0
    theta_1_temp=0
    theta_2_temp=0
    m=len(data_set)
    for i in range(0,m-1):
        theta_0_temp+=-(1)*(y[i]-((theta_1 * x[i] + theta_0+(theta_2*x1[i]))))
        theta_1_temp+=-(1)*(y[i]-((theta_1*x[i]+theta_0+(theta_2*x1[i]))))*x[i]
        theta_2_temp +=-(1)*(y[i]-((theta_1*x[i]+theta_0)+(theta_2*x1[i])))*x1[i]
    theta_0_new = theta_0 - (alpha * theta_0_temp)
    theta_1_new = theta_1 - (alpha * theta_1_temp)
    theta_2_new = theta_2 - (alpha * theta_2_temp)
    new_error=h_theta_avg_error(theta_0_new,theta_1_new,theta_2_new,data_set,m,x,x1,y)
    return(theta_0_new,theta_1_new,theta_2_new,new_error)
```

```python
import matplotlib.pyplot as plt
def plot(theta_0,theta_1,theta_2,data_set,m,x,x1,y):
    y_predict=[]
    y_list_2=[]
    #colors = (1)
    for i in range(0,m-1):
        y_predict.append((theta_0 + theta_1*x[i]+theta_2*x1[i]))    
       
    plt.scatter(x,x1)
    plt.plot(y_predict,'r-')
    plt.plot(y,'g-')
    plt.xlim((1,11))
```
